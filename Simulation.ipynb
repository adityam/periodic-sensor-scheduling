```julia
C = 3.0; initial=50.0; iterations=1_000; alpha=1.0;
```

```julia
sync = syncTrace(users, C; iterations=iterations, initial=initial, alpha=alpha);
sync[end-20:end,:]
```

```julia
async = asyncTrace(users, C; iterations=iterations, initial=initial, alpha=alpha)
async[end-20:end,:]
```

```julia
k = 100:500
```

```julia
plot(k, [sync[k,:lambda], async[k, :lambda]])
```

```julia
plot(k, sync[k,:MSE], label="sync", title="MSE")
plot!(k, async[k,:MSE], label="async")
```

```julia
plot(k, sync[k,:C], label="sync", title="Used Capacity")
plot!(k, async[k,:C], label="async")
```
